```python
# for given path, get length, sampling rate, and bit depth
import torchaudio
import os
import tqdm
def get_wav_info(path):
    # get length, sr, bit depth, and # of chanels
    info = torchaudio.info(path)
    return info.num_frames, info.sample_rate, info.bits_per_sample, info.num_channels

pth = '/mnt/arrakis_data/pnlong/lnac/youtube_mix/'



wav_info = {}
for root, _, files in tqdm.tqdm(os.walk(pth)):
    for f in files:
        if f.lower().endswith(('.flac', '.wav')):
            full_path = os.path.join(root, f)
            try:
                length, sr, bps, ch = get_wav_info(full_path)
            except Exception as e:
                print(f"Error reading {full_path}: {e}")
                continue
            key = os.path.basename(full_path)
            wav_info[key] = {'length': length, 'sample_rate': sr, 'bits_per_sample': bps, 'n_channels': ch}
            

# split 88-6-6
train_info = {}
val_info = {}
test_info = {}
for i, (k, v) in enumerate(wav_info.items()):
    if i % 100 < 88:
        train_info[k] = v
    elif i % 100 < 94:
        val_info[k] = v
    else:
        test_info[k] = v

# write to json
import json
with open("youtube_mix_train_info.json", "w") as f:
    json.dump(train_info, f)
with open("youtube_mix_val_info.json", "w") as f:
    json.dump(val_info, f)
```

```python
# make 90/10 train/val split of birdvox files
# it can't be fully random, as the data is underlying from 11 macro 12-hour recordings
# so the set of original recordings between train and val should be disjoint
import os
import random
import json

# file names are like: 'unit06_[ogfilename]_000.flac'

with open('birdvox_info.json', 'r') as f:
    wav_info = json.load(f)
all_files = list(wav_info.keys())

# group by original recording
recording_dict = {}
for f in all_files:
    rec_id = f.split('_')[0] + '_' + f.split('_')[1]  # e.g. unit06_[ogfilename]
    if rec_id not in recording_dict:
        recording_dict[rec_id] = []
    recording_dict[rec_id].append(f)

# try to split it as 90-10 as possible while keeping recordings disjoint
all_recordings = list(recording_dict.keys())
n_files_per_rec = {rec: len(files) for rec, files in recording_dict.items()}

random.shuffle(all_recordings)
train_files = []
val_files = []
train_recs = set()
val_recs = set()
total_files = len(all_files)
train_target = int(0.9 * total_files)
current_train_count = 0
for rec in all_recordings:
    rec_files = recording_dict[rec]
    if current_train_count + n_files_per_rec[rec] <= train_target:
        train_files.extend(rec_files)
        current_train_count += n_files_per_rec[rec]
        train_recs.add(rec)
    else:
        val_files.extend(rec_files)
        val_recs.add(rec)
print(f"Train files: {len(train_files)}, Val files: {len(val_files)}")
print(f"Train recordings: {train_recs}, Val recordings: {val_recs}")

# assert disjointness
assert train_recs.isdisjoint(val_recs), "Train and Val recordings are not disjoint!"
assert len(train_files) + len(val_files) == total_files, "File count mismatch!"
assert set(train_files).isdisjoint(set(val_files)), "Train and Val files are not disjoint!"

# make train info and val info jsons
train_info = {f: wav_info[f] for f in train_files}
val_info = {f: wav_info[f] for f in val_files}

# save
with open('birdvox_train_info.json', 'w') as f:
    json.dump(train_info, f)
with open('birdvox_val_info.json', 'w') as f:
    json.dump(val_info, f)
```

```python
# for every dataset, load train metadata and get total number of samples (that is length of each file * sample rate * number of bytes * number of files)
import tqdm
import random

# set seed
SEED = 42
random.seed(SEED)


SPLIT_RATIO = 0.99  # for datasets that don't have a val_metadata_path, we will do a random split of files based on this ratio

MIN_VAL_SAMPLES = 10_000_000

full_tokens_train = 0
full_tokens_val = 0
new_train_files = {}
new_val_files = {}
for dataset_name, dataset_info in all_datas["datasets"].items():
    print(f"Processing {dataset_name}")
    train_metadata_path = dataset_info["train_metadata_path"]
    train_metadata = json.load(open(train_metadata_path))
    total_samples = 0
    val_tokens = 0
    bit_depth_cap = dataset_info.get("bit_depth_cap", None)

    # if there isn't a val_metadata_path, then we have to partition the data based on the total number of samples, so we need to know the total number of samples in the training set
    # then we need to generate a split of random files that roughly gets to a desired split ratio (e.g. 90-10) based on the total number of samples in each file, rather than just the number of files, since some files are much longer than others
    samples_per_file = {}
    for file, metadata in train_metadata.items():
        length = metadata['length']
        sample_rate = metadata['sample_rate']
        bits_per_sample = metadata['bits_per_sample']
        bits_per_sample = min(bits_per_sample, bit_depth_cap) if bit_depth_cap is not None else bits_per_sample
        n_channels = metadata['n_channels']
        total_samples += length * (bits_per_sample / 8) * n_channels
        samples_per_file[file] = length * (bits_per_sample / 8) * n_channels
    
    if "val_metadata_path" not in dataset_info:
        print(f"\tNo val_metadata_path for {dataset_name}, doing {SPLIT_RATIO*100:.0f}-{(1-SPLIT_RATIO)*100:.0f} split based on samples per file")
        # find optimal split of files to get to 90-10 split based on samples_per_file
        new_train_files[dataset_name] = {}
        new_val_files[dataset_name] = {}
        files = list(samples_per_file.keys())
        random.shuffle(files)
        train_target = min(SPLIT_RATIO * total_samples, total_samples - MIN_VAL_SAMPLES)
        current_train_count = 0
        for f in files:
            if current_train_count + samples_per_file[f] <= train_target:
                current_train_count += samples_per_file[f]
                new_train_files[dataset_name][f] = train_metadata[f]
                
            else:
                val_tokens += samples_per_file[f]
                new_val_files[dataset_name][f] = train_metadata[f]
    else:
        print(f"\tVal metadata path found for {dataset_name}, calculating val tokens directly")
        val_metadata_path = dataset_info["val_metadata_path"]
        val_metadata = json.load(open(val_metadata_path))
        for file, metadata in val_metadata.items():
            length = metadata['length']
            sample_rate = metadata['sample_rate']
            bits_per_sample = metadata['bits_per_sample']
            bits_per_sample = min(bits_per_sample, bit_depth_cap) if bit_depth_cap is not None else bits_per_sample
            n_channels = metadata['n_channels']
            val_tokens += length * (bits_per_sample / 8) * n_channels

    total_samples = total_samples if "val_metadata_path" in dataset_info else current_train_count
    print(f"\t{dataset_name}: {total_samples/ 1e6:,.2f} M samples, {val_tokens/ 1e6:,.2f} M val samples")
    full_tokens_train += total_samples
    full_tokens_val += val_tokens

print(f"Total tokens across all datasets: {(full_tokens_train + full_tokens_val)/ 1e9:,.2f} B, Train tokens: {full_tokens_train/ 1e9:,.2f} B, Val tokens: {full_tokens_val/ 1e9:,.2f} B")
print(f"Val percentage: {full_tokens_val / (full_tokens_train + full_tokens_val) * 100:.2f}%")
```

```python
# write new_val_files to json, and separate jsons
for dataset_name, val_files in new_val_files.items():
    with open(f"{dataset_name}_new_val_info.json", "w") as f:
        json.dump(val_files, f)

with open("new_val_files.json", "w") as f:
    json.dump(new_val_files, f)

# same with train files
for dataset_name, train_files in new_train_files.items():
    with open(f"{dataset_name}_new_train_info.json", "w") as f:
        json.dump(train_files, f)
with open("new_train_files.json", "w") as f:
    json.dump(new_train_files, f)
```
